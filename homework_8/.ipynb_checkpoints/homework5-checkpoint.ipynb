## 1. Выделение когорт по времени присоединения

```python
#размер когорты сделаем один месяц
cohort_size = 30
#интервал времени между когортами для сравнения
interval_between_cohorts = 60

#найдем для пользоваателей их даты присоединения:
g_data = data.groupby('AuthorUserId')['CreationDateTimestamp'].first().reset_index()
#колонка delta - день присоединения с начала наблюдения:
g_data['delta'] = (g_data['CreationDateTimestamp'] - data['CreationDateTimestamp'].min()).apply(lambda x: x.days)


#получим айдишники  пользователей присоединившихся в первый месяц и во второй:
cohort_1_users = g_data[g_data['delta']<=cohort_size]['AuthorUserId'].unique()
cohort_2_users = g_data.loc[(g_data['delta']>interval_between_cohorts) &
                            (g_data['delta']<=interval_between_cohorts+cohort_size)]['AuthorUserId'].unique()

#полные логи событий для каждой из когорт:
data_c1 = data[data['AuthorUserId'].isin(cohort_1_users)].copy()
data_c2 = data[data['AuthorUserId'].isin(cohort_2_users)].copy()
```

Визуализируем теперь аналогично время регистрации новых событий/ новых пользователей для выделенных когорт:

```python
def visualize_users(df):
    data = df.copy()
    min_date = data['CreationDateTimestamp'].min()

    g_data = data.groupby('AuthorUserId').first().reset_index()
    g_data['delta'] = (g_data['CreationDateTimestamp'] - min_date).apply(lambda x: x.days/5)

    sns.set(style='white', context='notebook', rc={'figure.figsize':(8,5)})

    sns.distplot(g_data['delta'], kde=False, bins=10, norm_hist=False, label='')
    plt.xlabel('Месяцы с начала наблюдения когортой')

    plt.legend(labels=["Пользователи, пришедшие в этот месяц (интервал 5 дней)"])
    plt.xlim([0,10])
    plt.show()
    print('начало наблюдения за когортой: '+ str(min_date))

def visualize_creations(df):
    data = df.copy()
    min_date = data['CreationDateTimestamp'].min()
    max_date = data['CreationDateTimestamp'].max()

    data['delta'] = (data['CreationDateTimestamp'] - min_date).apply(lambda x: np.floor(x.days/30))

    sns.set(style='white', context='notebook', rc={'figure.figsize':(8,5)})
    bin_num=int(np.floor((max_date-min_date).days/30))
    
    sns.distplot(data['delta'], kde=False,bins=bin_num, norm_hist=False, label='')
    plt.xlabel('Месяцы с начала наблюдения за когортой')

    plt.legend(labels=["Все созданные ноутбуки этих пользователей"])
    plt.xlim([0,20])
    plt.show()
    print('начало наблюдения за когортой: '+ str(min_date))
```

```python
 visualize_users(data_c1)
```

Мы разбили график по пяти дням чтобы визуализировать месячную когорту и как внутри нее распределялся приток пользователей, видим что в самом начале был всплеск активности новых пользователей

```python
visualize_creations(data_c1)
```

Видим плавное замедление в течении года и всплеск актвности через год после первого созданного ноутбука

```python
visualize_creations(data_c2)
```

Вторая когорта быстрее теряла интерес к созданию ноутбуков, видимо пользователи пришедшие в первый месяц были сильнее заинтересованы в нашем продукте

Можем построить, как убывала доля активных пользователей в каждой из когорт интервал к интервалу с момента начала наблюденя за когортой:

```python
data_c1['delta'] = (data_c1['CreationDateTimestamp'] - data_c1['CreationDateTimestamp'].min())\
                    .apply(lambda x: np.floor(x.days / cohort_size))


data_c2['delta'] = (data_c2['CreationDateTimestamp'] - data_c2['CreationDateTimestamp'].min())\
                    .apply(lambda x: np.floor(x.days / cohort_size))
```

```python
retention_c1 =  data_c1.groupby('delta')['AuthorUserId'].apply(lambda x: len(set(x)))
retention_c2 =  data_c2.groupby('delta')['AuthorUserId'].apply(lambda x: len(set(x)))


sns.lineplot(x = retention_c1.index, y = retention_c1.values/max(retention_c1.values),
             lw = 2, marker="o", markersize = 8)

sns.lineplot(x = retention_c2.index, y = retention_c2.values/max(retention_c2.values),
             lw = 2, marker="o", markersize = 8)

plt.xlabel('периоды с начала наблюдения за когортой')
plt.ylabel('доля активных пользователей')

plt.legend(labels=["когорта 1","когорта 2"])
plt.show()
```

Мы видим, что когорта 1го месяца обладает сильно боолеее высоким retention, чем когорта пользователей 4-го месяца



Сравним эти группы по числу созданных ноутбуков на человека.

```python
creations_c1 =  data_c1.groupby('delta').apply(lambda x: len(x['Title'])/x['AuthorUserId'].nunique())
creations_c2 =  data_c2.groupby('delta').apply(lambda x: len(x['Title'])/x['AuthorUserId'].nunique())
```

```python
sns.lineplot(x = creations_c1.index, y = creations_c1.values,
             lw = 2, marker="o", markersize = 8)

sns.lineplot(x = creations_c2.index, y = creations_c2.values,
             lw = 2, marker="o", markersize = 8)

plt.xlabel('периоды с начала наблюдения за когортой')
plt.ylabel('число созданных ноутбуков')

plt.legend(labels=["когорта 1","когорта 2"])
plt.show()
```

Видно, что когорта 1 - создавала боольше ноутбуков на человека. Однако пользователи из второй когорты на 13м месяце продукта были более заинтересованы.


Построим сумарную оценку ноутбуков для каждой из когорт:

```python
conv_c1 =  data_c1.groupby('delta').apply(lambda x: (x['TotalVotes']).sum()/x['AuthorUserId'].nunique())
conv_c2 =  data_c2.groupby('delta').apply(lambda x: (x['TotalVotes']).sum()/x['AuthorUserId'].nunique())
# data_c1.head()
sns.lineplot(x = conv_c1.index, y = conv_c1.values,
             lw = 2, marker="o", markersize = 8)

sns.lineplot(x = conv_c2.index, y = conv_c2.values,
             lw = 2, marker="o", markersize = 8)

plt.xlabel('периоды с начала наблюдения за когортой')
plt.ylabel('суммарная оценка')

plt.legend(labels=["когорта 1","когорта 2"])
plt.show()
```

Видно, что работы пользователей когорты 1 чаще получают голоса от других пользователей

## 2. Выделение сегментов внутри когорт


Выделим сегменты по популярным авторам

```python
data['TotalVotes'].value_counts()
```

```python
#найдем для пользоваателей их даты присоединения:
g_data = data.groupby('AuthorUserId')['CreationDateTimestamp'].first().to_frame()
#колонка delta - день присоединения с начала наблюдения:
g_data['delta'] = (g_data['CreationDateTimestamp'] - data['CreationDateTimestamp'].min()).apply(lambda x: x.days)
```

```python
data['my_acquisition_day'] = data['AuthorUserId'].map(g_data['delta'].to_dict())
```

Разделим  когорты на сегменты по авторам с голосами на их работах больше 5 и меньше 5

```python
cohort_size =  30

cohort_1_users = data.loc[(data['my_acquisition_day']<=cohort_size) &
                          (data['TotalVotes']>=5)]['AuthorUserId'].unique()

cohort_2_users = data.loc[(data['my_acquisition_day']<=cohort_size) &
                          (data['TotalVotes']<5)]['AuthorUserId'].unique()

#полные логи событий для каждой из когорт:
data_c1 = data[data['AuthorUserId'].isin(cohort_1_users)].copy()
data_c2 = data[data['AuthorUserId'].isin(cohort_2_users)].copy()
```

Далее можем провести сравнительный анализ аналогичный проделанному выше по выделенным  когортам:

```python
data_c1['delta'] = (data_c1['CreationDateTimestamp'] - data_c1['CreationDateTimestamp'].min())\
                    .apply(lambda x: np.floor(x.days / cohort_size))


data_c2['delta'] = (data_c2['CreationDateTimestamp'] - data_c2['CreationDateTimestamp'].min())\
                    .apply(lambda x: np.floor(x.days / cohort_size))
```

```python
retention_c1 =  data_c1.groupby('delta')['CreationDateTimestamp'].apply(lambda x: len(set(x)))
retention_c2 =  data_c2.groupby('delta')['CreationDateTimestamp'].apply(lambda x: len(set(x)))


sns.lineplot(x = retention_c1.index, y = retention_c1.values/max(retention_c1.values),
             lw = 2, marker="o", markersize = 8)

sns.lineplot(x = retention_c2.index, y = retention_c2.values/max(retention_c2.values),
             lw = 2, marker="o", markersize = 8)

plt.xlabel('периоды с начала наблюдения за когортой')
plt.ylabel('доля активных пользователей')

plt.legend(labels=["когорта votes > 5","когорта votes < 5"])
plt.show()
```

Видно, что пользователи которые имели одобрение чаще возвращались


## 3.  Когортная матрица retention для многих групп:


```python
data['delta'] = (data['CreationDateTimestamp']-data['CreationDateTimestamp'].min()).apply(lambda x: x.days)
```

```python
#за интервал возьмем 1 месяц:
cohort_size = 30

data['acquisition_period'] = data['acquisition_day'].apply(lambda  x: np.floor(x / cohort_size))
data['activity_week'] = data['delta'].apply(lambda  x: np.floor(x / cohort_size))
data['lifetime_period'] =  data['activity_week'] - data['acquisition_period']
```

```python
cohorts = data.groupby(['acquisition_period','lifetime_period']).agg({'AuthorUserId':'nunique'}).reset_index()

inital_users_count = cohorts[cohorts['lifetime_period'] == 0][['acquisition_period','AuthorUserId']]
inital_users_count = inital_users_count.rename(columns={'AuthorUserId':'cohort_users'})


cohorts = cohorts.merge(inital_users_count,on='acquisition_period')
cohorts['metric'] = cohorts['AuthorUserId']/cohorts['cohort_users']

cohort_metric_pivot = cohorts.pivot_table(index='acquisition_period',
                                      columns='lifetime_period'
                                      ,values='metric',
                                      aggfunc='sum')

sns.set(style='white', context='notebook', rc={'figure.figsize':(16,10)})

plt.title('Cohorts: Percent of Active Users')
sns.heatmap(cohort_metric_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', cmap='viridis',
            vmin=0.09, vmax=0.45,)

b, t = plt.ylim() 
b += 0.5 
t -= 0.5 
plt.ylim(b, t)

plt.show();
```

Действительно наблюдаем долгоиграющий интерес у когорты первого месяца. Когорта 13го месяца и немного 14го тоже показывают силу, скорее всего связано с проведение контеста как и предполагалось ранее
