Наблюдаем неуказанный числовой параметр, номер счета, код акции, количество, дата выставления счета, цена за единицу, пользовательский id. Возможно это данные о покупках акций на фондовом рынке

Выясним временной промежуток проведения эксперимента

```python
dataA['InvoiceDate'] = pd.to_datetime(dataA['InvoiceDate'])
min_date = dataA['InvoiceDate'].min()
max_date = dataA['InvoiceDate'].max()
print(min_date, max_date)
```

```python
dataB['InvoiceDate'] = pd.to_datetime(dataB['InvoiceDate'])
min_date = dataB['InvoiceDate'].min()
max_date = dataB['InvoiceDate'].max()
print(min_date, max_date)
```

Непривычно что стартовая дата у данных сильно отличается, по всей видимости B вариант это базовый вариант который был до теста, вариант A начал тестироваться 31 марта  
  
Преобразуем данные для построения таблицы retention по когортам присоединения пользователей в каждый месяц

```python
#найдем для пользоваателей их даты присоединения:
g_dataA = dataA.groupby('CustomerID')['InvoiceDate'].first().to_frame()
#колонка delta - день присоединения с начала наблюдения:
g_dataA['delta'] = (g_dataA['InvoiceDate'] - dataA['InvoiceDate'].min()).apply(lambda x: x.days)

#найдем для пользоваателей их даты присоединения:
g_dataB = dataB.groupby('CustomerID')['InvoiceDate'].first().to_frame()
#колонка delta - день присоединения с начала наблюдения:
g_dataB['delta'] = (g_dataB['InvoiceDate'] - dataB['InvoiceDate'].min()).apply(lambda x: x.days)
```

```python
dataA['acquisition_day'] = dataA['CustomerID'].map(g_dataA['delta'].to_dict())
dataB['acquisition_day'] = dataB['CustomerID'].map(g_dataB['delta'].to_dict())
```

```python
rete_period = 30 #выбираем период в 30 дней - примерно месяц
dataA['delta'] = (dataA['InvoiceDate']-dataA['InvoiceDate'].min()).apply(lambda x: x.days)
dataA['acquisition_period'] = dataA['acquisition_day'].apply(lambda  x: np.floor(x / rete_period))
dataA['activity_month'] = dataA['delta'].apply(lambda  x: np.floor(x / rete_period))
dataA['lifetime_period'] =  dataA['activity_month'] - dataA['acquisition_period']
```

```python
cohorts = dataA.groupby(['acquisition_period','lifetime_period']).agg({'CustomerID':'nunique'}).reset_index()

inital_users_count = cohorts[cohorts['lifetime_period'] == 0][['acquisition_period','CustomerID']]
inital_users_count = inital_users_count.rename(columns={'CustomerID':'cohort_users'})


cohorts = cohorts.merge(inital_users_count,on='acquisition_period')
cohorts['metric'] = cohorts['CustomerID']/cohorts['cohort_users']

cohort_metric_pivot = cohorts.pivot_table(index='acquisition_period',
                                      columns='lifetime_period'
                                      ,values='metric',
                                      aggfunc='sum')

sns.set(style='white', context='notebook', rc={'figure.figsize':(16,10)})

plt.title('Процент активных пользователей')
sns.heatmap(cohort_metric_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', cmap='viridis',
            vmin=0.09, vmax=0.5,)

b, t = plt.ylim() 
b += 0.5 
t -= 0.5 
plt.ylim(b, t)

plt.show();
```

```python
rete_period = 30
dataB['delta'] = (dataB['InvoiceDate']-dataB['InvoiceDate'].min()).apply(lambda x: x.days)
dataB['acquisition_period'] = dataB['acquisition_day'].apply(lambda  x: np.floor(x / rete_period))
dataB['activity_month'] = dataB['delta'].apply(lambda  x: np.floor(x / rete_period))
dataB['lifetime_period'] =  dataB['activity_month'] - dataB['acquisition_period']
```

```python
cohorts = dataB.groupby(['acquisition_period','lifetime_period']).agg({'CustomerID':'nunique'}).reset_index()

inital_users_count = cohorts[cohorts['lifetime_period'] == 0][['acquisition_period','CustomerID']]
inital_users_count = inital_users_count.rename(columns={'CustomerID':'cohort_users'})


cohorts = cohorts.merge(inital_users_count,on='acquisition_period')
cohorts['metric'] = cohorts['CustomerID']/cohorts['cohort_users']

cohort_metric_pivot = cohorts.pivot_table(index='acquisition_period',
                                      columns='lifetime_period'
                                      ,values='metric',
                                      aggfunc='sum')

sns.set(style='white', context='notebook', rc={'figure.figsize':(16,10)})

plt.title('Cohorts: Percent of Active Users')
sns.heatmap(cohort_metric_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', cmap='viridis',
            vmin=0.09, vmax=0.5,)

b, t = plt.ylim() 
b += 0.5 
t -= 0.5 
plt.ylim(b, t)

plt.show();
```

Наблюдаем что ретеншн пользователей первого месяца на протяжении года держится на очень хорошем уровне, видимо это месяц запуска приложения и мы получили лояльных пользователей которые давно ждали наш продукт  
  
По тепловой карте видно что в группе B retention выше чем в группе A

Попробуем посчитать средний retention двух групп и сравнить их на одном графике. Выберем 7 месяцев с начала наблюдения, так как значения последнего месяца нам будет некоректно использовать так как месяц не полный, при суммирований не будем считать декабрь

```python
dataB2 = dataB.loc[(dataB['acquisition_period'] != 0) & (dataB['lifetime_period'] <= 7)] #убираем первый месяц
dataA2 = dataA[dataA['lifetime_period'] <= 7] 

retentionA =  dataA2.groupby('lifetime_period')['CustomerID'].apply(lambda x: len(set(x)))
retentionB =  dataB2.groupby('lifetime_period')['CustomerID'].apply(lambda x: len(set(x)))


sns.lineplot(x = retentionA.index, y = retentionA.values/max(retentionA.values),
             lw = 2, marker="o", markersize = 8)

sns.lineplot(x = retentionB.index, y = retentionB.values/max(retentionB.values),
             lw = 2, marker="o", markersize = 8)

plt.xlabel('Месяц')
plt.ylabel('Retention')

plt.legend(labels=["группа A","группа B"])
plt.show()
```

### 2) Проведем сравнение survival curves

Для этого расчитаем для каждого события дельту от первого события пользователя. После этого удалим все события больше месяца (так как мы рассматриваем интервал в месяц). Расчитаем дельту последнего события для каждого пользователя и построим график по долям

```python
#найдем для пользоваателей их даты присоединения:
s_dataA = dataA.groupby('CustomerID')['InvoiceDate'].first().to_frame()
#найдем для пользоваателей их даты присоединения:
s_dataB = dataB.groupby('CustomerID')['InvoiceDate'].first().to_frame()

dataA['FirstDate'] = dataA['CustomerID'].map(s_dataA['InvoiceDate'].to_dict())
dataA['ReturnTime'] = dataA['InvoiceDate']-dataA['FirstDate']

dataB['FirstDate'] = dataB['CustomerID'].map(s_dataB['InvoiceDate'].to_dict())
dataB['ReturnTime'] = dataB['InvoiceDate']-dataB['FirstDate']

listDays = []
days = []
for i in range(0, 31):
    days.append(i)
    listDays.append(pd.Timedelta("{} days".format(i)))
    
dataA = dataA[(dataA['ReturnTime']) < pd.Timedelta("30 days")]
dataB = dataB[(dataB['ReturnTime']) < pd.Timedelta("30 days")]
    
#найдем для пользоваателей их последнюю сессию в течении месяца от присоединения:
returnA = dataA.groupby('CustomerID')['ReturnTime'].max().to_frame()
returnB = dataB.groupby('CustomerID')['ReturnTime'].max().to_frame()

usersA = returnA.size
usersByDaysA = []
usersB = returnB.size
usersByDaysB = []

for day in listDays:
    dayReturnA = returnA[(returnA['ReturnTime']) >= day]
    usersByDaysA.append(dayReturnA.size/usersA)
    
    dayReturnB = returnB[(returnB['ReturnTime']) >= day]
    usersByDaysB.append(dayReturnB.size/usersB)
    
sns.lineplot(x = days, y = usersByDaysA,
             lw = 2, marker="o", markersize = 8)

sns.lineplot(x = days, y = usersByDaysB,
             lw = 2, marker="o", markersize = 8)
```

Как мы видим в группе B пользователи теряются менее активно. Значит состояние продукта в группе B было эффективнее
  
Продолжать тест не нужно. Группа B показывает себя лучше
